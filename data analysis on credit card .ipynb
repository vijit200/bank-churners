# numerical variable

```python
numerical_feature=[feature for feature in df.columns if df[feature].dtype!='O']
```

```python
df[numerical_feature].head()
```

```python
discreate_feature=[feature for feature in numerical_feature if len(df[feature].unique())<20 and df[feature].dtype!='float' ]
```

```python
df[discreate_feature].head()
```

```python
df['Attrition_Flag'].unique()
```

```python
dicto={'Existing Customer':1,'Attrited Customer':0}
```

```python
df['Attrition_Flag']=df['Attrition_Flag'].map(dicto)
```

```python
df.head()
```

```python
#now checking relation between discreate and target attribute
```

```python
import matplotlib.pyplot as plt
%matplotlib.inline
for feature in discreate_feature:
    data=df.copy()
    data.groupby(feature)['Attrition_Flag'].mean().plot(kind='barh')
    plt.xlabel(feature)
    plt.title('v/s graph')
    plt.show()
```

```python
for feature in discreate_feature:
    data=df.copy()
    data.groupby(feature)['Attrition_Flag'].mean().plot.bar()
    plt.xlabel(feature)
    plt.title('v/s graph')
    plt.show()
```

```python
#here we see there is some relation b/w them 
```

# continous variable

```python
continous_feature=[feature for feature in numerical_feature if feature not in discreate_feature]
```

```python
df[continous_feature].head()
```

  # Q-Q Plot And Hist plot for continous variable
  q-q plot is used  to check the distribution is normal distribution  is or not

```python
import scipy.stats as stats
for feature in continous_feature:
    data=df.copy()
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    data[feature].hist(bins=25)
    plt.subplot(1,2,2)
    stats.probplot(df[feature],dist='norm',plot=plt)
    plt.xlabel(feature)
    plt.show()
```

```python
# as we see most of the data are skwed and distributed data
```

# outlier of numerical variable
